```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Load the data
file_path = '/content/Book1.csv'  # Update this if needed
data = pd.read_csv(file_path)
```

```python
# Define the features (X) and target (Y)
X = data[['X']]
Y = data['Y']
```

```python
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
# Initialize the Logistic Regression model
model = LogisticRegression()
```

```python
# Train the model
model.fit(X_train, Y_train)
```

```python
  # Make predictions
Y_pred = model.predict(X_test)
```

```python
# Plot the results
plt.scatter(X_test, Y_test, color='green', label='Actual Data')
plt.scatter(X_test, Y_pred, color='orange', label='Predicted Data', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Logistic Regression (Test Data)')
plt.show()
```

```python
# Print model parameters
print(f'Intercept: {model.intercept_[0]}')
print(f'Coefficients: {model.coef_[0][0]}')
```

```python
# Evaluate the model
accuracy = accuracy_score(Y_test, Y_pred)
conf_matrix = confusion_matrix(Y_test, Y_pred)
print(f'Accuracy: {accuracy}')
print(f'Conf_matrix:{conf_matrix}')
```
