```python
import pandas as pd
import numpy as np

# Step 1: Generate Sample Datasets
np.random.seed(1)
data_A = np.random.normal(loc=50, scale=5, size=100)   # Low spread
data_B = np.random.normal(loc=50, scale=15, size=100)  # High spread

# Save to CSV
pd.DataFrame({'Value': data_A}).to_csv('/content/dataset_A.csv', index=False)
pd.DataFrame({'Value': data_B}).to_csv('/content/dataset_B.csv', index=False)

# Load datasets
df_A = pd.read_csv('dataset_A.csv')
df_B = pd.read_csv('dataset_B.csv')

# Step 2: Compute Standard Deviation
std_A = df_A['Value'].std()
std_B = df_B['Value'].std()

print(f"Standard Deviation of Dataset A: {std_A:.2f}")
print(f"Standard Deviation of Dataset B: {std_B:.2f}")

# Step 3: Compute Covariance
# Merge datasets into a single DataFrame
df_combined = pd.DataFrame({'A': df_A['Value'], 'B': df_B['Value']})
cov_matrix = df_combined.cov()
covariance = cov_matrix.loc['A', 'B']

print(f"Covariance between Dataset A and B: {covariance:.2f}")

# Step 4: Justification
if std_A > std_B:
    print("Dataset A has a larger spread.")
else:
    print("Dataset B has a larger spread.")
```
