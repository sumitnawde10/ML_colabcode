```python
import json
import numpy as np

# Load dataset (Ensure mcp_dataset.json is uploaded in Colab)
dataset_path = "/content/mcp_dataset.json"
with open(dataset_path, "r") as f:
    dataset = json.load(f)

# Implementing AND Gate
def test_logic_gate(gate_name, weights, threshold):
    print(f"\nTesting {gate_name} Gate:")
    for i in range(len(dataset[gate_name]["X1"])):
        inputs = np.array([dataset[gate_name]["X1"][i]])
        if "X2" in dataset[gate_name]:  # Check if the gate needs a second input
            inputs = np.append(inputs, dataset[gate_name]["X2"][i])
        output = mcp_neuron(inputs, weights, threshold)
        print(f"Input: {inputs}, Output: {output}")
        print(f"Inputs for {gate_name}: {inputs}, Weights: {weights}, Threshold: {threshold}")


# McCulloch-Pitts Neuron Function
def mcp_neuron(inputs, weights, threshold):
    net_input = np.dot(inputs, weights)
    return 1 if net_input >= threshold else 0

# Define weights and thresholds for different gates
gate_params = {
    "AND": {"weights": [1, 1], "threshold": 2},
    "OR": {"weights": [1, 1], "threshold": 0.5},
    "NOT": {"weights": [-1], "threshold": -0.5},
}

# Run tests for each logic gate
for gate, params in gate_params.items():
    test_logic_gate(gate, params["weights"], params["threshold"])


```
