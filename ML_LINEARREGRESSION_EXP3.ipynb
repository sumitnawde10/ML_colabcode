```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#  Load dataset from CSV
file_path = '/content/linear_regression_dataset.csv'  # Update this if needed
data = pd.read_csv(file_path)
X = data[['X']]
Y = data[['Y']]
```

```python
#  Split dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#  Train a linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#  Make predictions on test data
Y_pred = model.predict(X_test)
```

```python
#  Plot regression against actual test data
plt.scatter(X_test, Y_test, color='green', label='Actual Data')
plt.plot(X_test, Y_pred, color='orange', linewidth=2, label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression (Test Data)')
plt.show()
```

```python
# Output Analysis
print(f'Scikit-learn Intercept: {model.intercept_[0]}')
print(f'Scikit-learn Slope: {model.coef_[0][0]}')
```

```python
#  Calculate error and accuracy
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)  # R-squared score for model accuracy
print(f'Mean Squared Error: {mse}')
print(f'R-squared Score (Accuracy): {r2}')
```
